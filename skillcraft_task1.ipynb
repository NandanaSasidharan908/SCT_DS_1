```python
import pandas as pd

df = pd.read_csv("1.1 - India_Historical_Population_Density_Data.csv")
df.head()
```

```python
df.columns = [col.strip().replace(" ", "_") for col in df.columns]
df.rename(columns={"popu1ation_growth_rate": "Population"}, inplace=True)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.bar(df["year"], df["Population_Density"], color="skyblue", edgecolor="black")
plt.title("Population Density Over Time")
plt.xlabel("Year")
plt.ylabel("Population Density")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 5))
plt.hist(df["Population_Density"], bins=10, color="lightgreen", edgecolor="black", alpha=0.7)
plt.title("Distribution of Population Density")
plt.xlabel("Population Density")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
